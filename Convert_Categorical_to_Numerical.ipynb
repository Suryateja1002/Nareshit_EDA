# **map method**
- before applying map method first get the unique labels of the column
- for example case_status is a categorical column
- it has two unique labels they are
    - Denied and
    - Certified
- Create a dictionary 
    - key as label
    - valu as number
- d={'Certified':0,'Denied':1}
- The dictionary we need to map the case_status column

```python
visa_df['case_status'].unique()
```

```python
# map method
d={'Certified':0,'Denied':1}
#overriding is best option
visa_df['case_status']=visa_df['case_status'].map(d)
visa_df
```

```python
# continent column
# use map method
d1={}
unique_labels=visa_df['continent'].unique()
for i in range(len(unique_labels)):
    d1[unique_labels[i]]=i
visa_df['continent']=visa_df['continent'].map(d1)
visa_df
```

# ***np.where***
- It is applicable for only binary(two) labels
- in case status column we have only two labels 'Certified' and 'Denied'
- If case_status == 'Certified' replace that as 0, otherwise 1
- np.where requires three arguments
    - condition
    - True
    - False


```python
np.where(visa_df['case_status'] == 'Certified',1,0)
```

# ****onehot encoder****
- It means at a time one will ON and other will OFF
- In case status column we have only two labels 'Certified' and 'Denied'
- when we apply onehot encoder on 'case_status' it will create two more extra columns 
    - case_status_certified
    - case_status_denied
 
  
|case_status|  case_status_certified |   case_status_denied
|-|-|-|
|certified|
|denied|


- ----------------------------------------------------------------------------------------------------
- ***Advantage of using onehot encoder***
    - if we develop an ML model it is very important thee columns should be independent to each other
    - so by using onehot encoder we can create for case_status  two extra columns
    - which are independent to each other
    - which means the row values at a time only one column has value 1.
    - Columns are independent to each other
    - which means **90 degrees phase shift**
    - which means **Perpendicular to each other**
    - which means **Orthogonal to each other**
- -------------------------------------------------------------------------------------------------------
- ***Disadvantage of using onehot encoder***
    - If a column has 100 unique labels , onehot encoder will create 100 new columns
    - The data will became sparse, which means huge
    - Columns are more meaning Dimensions are more
    - So that **Processing time** will be more
    - The **Memory Consumption** will be more
    - leads to
    - **Curse of Dimensionality**
- --------------------------------------------------------------------------------------------------------
