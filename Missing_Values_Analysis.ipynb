# **Method-4**
- **Mean**
- **Mode**
- **Median**

```python
age_mean=data1['Age'].mean()
age_mean
```

```python
# instead of providing a random number
# we are filling with mean of the data
data1['Age'].fillna(age_mean,inplace=True)
data1
```

```python
dict1={'Names':['Ramesh','Suresh',np.nan,'Mahesh'],
      'Age':[31,32,33,np.nan],
      'City':[np.nan,'Hyd','Mumbai','Chennai']}
data1=pd.DataFrame(dict1)
data1
```

```python
age_median=data1['Age'].median()
data1['Age'].fillna(age_median,inplace=True)
data1
```

```python
dict1={'Names':['Ramesh','Suresh',np.nan,'Mahesh'],
      'Age':[31,32,33,np.nan],
      'City':[np.nan,'Hyd','Mumbai','Chennai']}
data1=pd.DataFrame(dict1)
data1
```

```python
age_mode=data1['Age'].mode()
data1['Age'].fillna(age_mode,inplace=True)
data1
```

# **Method-5**

# **KNN imputer**
- KNN - K- nearest neighbour
- in knn instead of taking mean of all the values
- we will choose neighbour data
- we will take those mean only

```python
from sklearn.impute import KNNImputer
knn=KNNImputer(n_neighbors=2)
knn.fit_transform(data1[['Age']])
```
