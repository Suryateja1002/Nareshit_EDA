```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file1=r"C:\Users\surya\Downloads\Python files\Visadataset.csv"
visa_df=pd.read_csv(file1)
visa_df
```

```python
prevail_wage=visa_df['prevailing_wage']
prevail_wage
median=np.median(visa_df['prevailing_wage'])
# step_1
l=[]
Q1= np.percentile(visa_df['prevailing_wage'],25)
Q2= np.percentile(visa_df['prevailing_wage'],50)
Q3= np.percentile(visa_df['prevailing_wage'],75)

# step-2
IQR = Q3 - Q1 

# step-3
lb= Q1 - (1.5 * IQR)
up= Q3 + (1.5 * IQR)
for i in prevail_wage:
    if i<lb or i>up:
        l.append(median)
    else:
        l.append(i)
#print(l)
print(len(l))

# adding ba column after replacing outliers with median value
visa_df['prevailing_wage_new']=l
visa_df
```

```python
plt.boxplot(visa_df['prevailing_wage_new'])
plt.show()
```

# **np.where** (Important)
- **np.where(con,True value,False value)**
- above  replace one we use traditional method
- by using for loop, list and if-else
- instead of that we can use np.where

```python
dict1={'Names':['a','b','c','d'], 'Num':[1,2,3,4]}
d1=pd.DataFrame(dict1)
d1
```

```python
# we want to replace num which 100 has >2
# otherwise keep same number
l=[]
for i in d1['Num']:
    if i>2:
        l.append(100)
    else:
        l.append(i)
d1['Num']=l
d1
```

```python
l=np.where(d1['Num']>2 , 100 ,d1['Num'])
d1['Num']=l
d1
```

```python
median=np.median(visa_df['prevailing_wage'])
# step_1
Q1= np.percentile(visa_df['prevailing_wage'],25)
Q2= np.percentile(visa_df['prevailing_wage'],50)
Q3= np.percentile(visa_df['prevailing_wage'],75)

# step-2
IQR = Q3 - Q1 

# step-3
lb= Q1 - (1.5 * IQR)
up= Q3 + (1.5 * IQR)
c1=visa_df['prevailing_wage']<lb
c2=visa_df['prevailing_wage']>up
c=c1|c2
l=np.where(c, median , visa_df['prevailing_wage'])

visa_df['prevailing_wage_new']=l
visa_df
plt.boxplot(visa_df['prevailing_wage_new'])
plt.show()
```
