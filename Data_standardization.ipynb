- Step-1:- Take prevailing_wage column
-    - z-score= x-mean/sigma
- Step-2:- Calculate mean of prevailing_wage
- Step-3:- Calculate Standard Deviation of prevailing_wage
- Step-4:- Numerator=prevailing_wage-mean
- Step-5: Prevailing_wage_Zscore= Numerator/Denominator

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file1=r"C:\Users\surya\Downloads\Python files\Visadataset.csv"
visa_df=pd.read_csv(file1)
visa_df
```

```python
prevailing_wage=visa_df['prevailing_wage']
mean=np.mean(visa_df['prevailing_wage'])
std=np.std(visa_df['prevailing_wage'])
#prevailing_wage_zscore=(prevailing_wage-mean) / std
visa_df['prevailing_wage_zscore']=(prevailing_wage-mean) / std
visa_df
```

# **Normalization**

```python
# Normalization
file1=r"C:\Users\surya\Downloads\Python files\Visadataset.csv"
visa_df=pd.read_csv(file1)
prevailing_wage=visa_df['prevailing_wage']
prevailing_wage_min=visa_df['prevailing_wage'].min()
prevailing_wage_max=visa_df['prevailing_wage'].max()
visa_df['prevailing_wage_minmaxscalar']=(prevailing_wage - prevailing_wage_min) / (prevailing_wage_max - prevailing_wage_min)
visa_df
```

```python
visa_df['prevailing_wage_minmaxscalar'].min(),visa_df['prevailing_wage_minmaxscalar'].max()
```

```python
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
visa_df['prevailing_wage_mms']=mms.fit_transform(visa_df[['prevailing_wage']])
visa_df['prevailing_wage_mms']
visa_df
```
