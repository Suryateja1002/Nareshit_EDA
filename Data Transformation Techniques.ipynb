# **Data Transformation Techniques**
- Generally used for to convert Normal distribution
- Because all statistical math analysis by assumption Data follows Normal distribution
- It is also avoid skewness 
- We have some important transformation
    - Log transformation
    - Exponential transformation
    - Reciprocal transformation
    - Square root transformation
    - Power transformaton

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Exponential_data

```python
exp_data=np.random.exponential(size=10000)

# We are consider exponential data set with random 1k observations
# using numpy package
exp_data[:20]
```

```python
# Histogram
plt.hist(exp_data,bins=100,label='Exponential')
plt.legend()
plt.show()
```

```python
# Normal_data
norm_data=np.random.normal(size=10000)
plt.hist(norm_data,bins=100,label='Normal')
plt.legend()
plt.show()
```

# **Log_Transformation**
- np.log is used for log transformation
- 
Generally log transformation will not convert data into norma
- It avoids skewness
- s
np.log means natural logoritto the hm base=e

```python
np.log(2)
```

```python
log_data=np.log(exp_data)
log_data[:20]
```

```python
plt.hist(log_data,bins=100,label='Logarithmic')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1).hist(exp_data,bins=100,label='Exponential')
plt.legend()
plt.subplot(1,2,2).hist(log_data,bins=100,label='Logarithmic')
plt.legend()
plt.show()
```

# **Reciprocal_Transformation**

- Reciprocal transformation fails when data has zero value
- Take the exponential data-  rec_data= 1/exp_dat

```python
np.reciprocal(3)
```

```python
rec_data=np.reciprocal(exp_data)
rec_data[:20]
```

```python
exp_data[:20],rec_data[:20]
```

# **Squareroot_Transformation**

```python
print(25**2)
print(25**(1/2))
print(np.sqrt(25))
```

```python
sqrt_data=np.sqrt(exp_data)
sqrt_data[:20]
```

```python
plt.hist(sqrt_data,bins=100,label='Squareroot')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1).hist(exp_data,bins=100,label='Exponential')
plt.legend()
plt.subplot(1,2,2).hist(sqrt_data,bins=100,label='Squareroot')
plt.legend()
plt.show()
```

```python
import seaborn as sns
sns.displot(sqrt_data)
```

# **Power_Transformation**
- It is related to sklearn package- 
Package name:sklearn.preprocessing
- 
Method nam-ePower Transformer
- Inside Box_Cox , yeo-jhonsonson

```python
exp_data
```

```python
from sklearn.preprocessing import PowerTransformer
pt=PowerTransformer()
power_data=pt.fit([exp_data]).transform([exp_data])
power_data
```

```python
plt.hist(power_data[0],label='PowerTransform')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(2,2,1).hist(exp_data,bins=100,label='Exponential')
plt.legend()
plt.subplot(2,2,2).hist(log_data,bins=100,label='Logarithmic')
plt.legend()
plt.subplot(2,2,3).hist(sqrt_data,bins=100,label='Squareroot')
plt.legend()
plt.subplot(2,2,4).hist(power_data[0],label='PowerTransform')
plt.legend()
plt.show()
```
