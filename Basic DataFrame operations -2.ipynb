# to get a specific row
- head will give top n rows
- tail will give bottom n rows

# take , loc , iloc
- take([]) ---> we need give value in [] or () inside df.take()

```python
df.take([2])
```

```python
df.take([2,3,4,5,6])
```

```python
df.take((2))
```

```python
df.take((2,3,4,5,6))
```

# Important

```python
# row ----> 100,200, 300 Col-------> 4,8,11
df.take([100,200,300]).take([4,8,11],axis=1)
```

# Drawback in take()
- take don't take rows and columns at a time
- so we use iloc()

# iloc[]
- iloc takes [] brackets not()
- df.iloc[<rows>,<columns>]
- df.iloc[<start:end>,<columns[start:end]>]
- df.iloc[rows,columns]

```python
# specific rows and specific columns
df.iloc[2:5,4:6]
```

```python
# all rows and specific columns
df.iloc[:,2:5]
```

```python
# all columns and specified rows
df.iloc[2:5]
```

# Important

```python
# row ----> 100,200, 300 Col-------> 4,8,11
df.iloc[[100,200,300],[4,8,11]]
```

- No brackets   -------> series
- brackets  ------------> dataframe


```python
df.iloc[[100,200,300],8]
```

```python
df.iloc[[100,200,300],[8]]
```

# Drawback of iloc
- if we want a specific column we need to use loc with that column name in the list

# loc

```python
df.loc[[100,200,300],['prevailing_wage']]
```
