# pd.DataFrame is used to create a data frame

```python
# create dataframe using list
Names=['Ramesh','Suresh','Satesh']
age=[25,30,35]

# to read a dataframe we use pd.DataFrame

pd.DataFrame(zip(Names,age))
```

# adding columns names by using columns in Dataframe 

```python
Names=['Ramesh','Suresh','Satesh']
age=[25,30,35]

# adding column names

pd.DataFrame(zip(Names,age),columns=['Names','age'])
```

# adding row names using index in dataframe

```python
Names=['Ramesh','Suresh','Satesh']
age=[25,30,35]

# adding row names

pd.DataFrame(zip(Names,age),columns=['Names','age'],index=['A','B','C'])
```

```python
Names=['Ramesh','Suresh','Satesh']
age=[25,30,35]

df=pd.DataFrame(zip(Names,age),columns=['Names','age'],index=['A','B','C'])
df
```

# adding a new column 
- df['New_column'] = column_list_name
- you need to have a list which is having some elements and that elements need to be equal to no of rows
- city_names=['Hyd','Bangluru','chennai']

```python
Names=['Ramesh','Suresh','Satesh']
age=[25,30,35]
city_names=['Hyd','Bangluru','chennai']
df=pd.DataFrame(zip(Names,age),columns=['Names','age'],index=['A','B','C'])
df['city']=city_names
df
```

# change the existing columns(updating the existing column)
- if you want to create a new column or update the new column both have same ways


```python
df['Names']=['Surya','Swamy','Raghu']
df
```

# dropping a column from a dataframe
- Inorder to drop a column
- we need to use drop method
- we need to give three parameters
    - drop column or row
    - mention the column name
    - Axis:-
    - Axis=0 reference as row
    - Axis=1 reference as column
    - you want to create a new dataframe or you want to override the existing dataframe
        - inplace=True

```python
df.drop('city',axis=1,inplace=True)
df
```

# dropping a row

```python
df.drop('C',axis=0,inplace=True)
df
```
